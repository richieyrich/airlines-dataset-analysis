# Airlines dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 15)

# read data
df = pd.read_csv('cleaned_air.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# most landed airlines

landing_counts_by_airline = df.groupby(['Operating Airline' ])['Landing Count'].sum()
print(landing_counts_by_airline)
# 5 most landed
top5 = landing_counts_by_airline.nlargest(24)
top5
```

```python
# plot a barchart

plt.figure(figsize=(15, 6))
sns.barplot(x= top5.index,y=top5.values )
plt.title('most landed airlines')
plt.xticks(rotation=45)
plt.xlabel('Landing Count')
plt.ylabel('Count')
plt.show()
```

scatterplot bw total land weight and count

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(x='Total Landed Weight', y='Landing Count', data=df)
plt.title('scatterplot bw total land weight and count')
plt.xlabel('Total Land Weight')
plt.ylabel('Landing Count')
plt.show
```

plot no. of aircrafts by region

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='GEO Region', hue= 'Aircraft Body Type', data=df, order=df['GEO Region'].value_counts().index)
plt.title('no. of aircrafts by region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='GEO Region', y= 'Landing Count', data=df)
plt.title('Landing Count of aircrafts by region')
plt.xlabel('Region')
plt.ylabel('Landing Count')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='Landing Aircraft Type',  data=df)
plt.title('no. of aircrafts by region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```
